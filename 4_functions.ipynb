## 1. Функції в математиці

У математиці функції &mdash; такі правила, що кожному вхідному значенню ставлять у відповідність одне вихідне значення. Наприклад, функція $f(x) = |x|$ є функцією, що кожному числу (дійсному чи навіть комплексному) ставить у відповідність його абсолютне значення.

У Python функції виконують схожу роль: отримують певні вхідні значення, виконують перетворення та ставлять у відповідність вихідне значення. Роль тієї самої математичної функції "модуль числа" виконує функція `abs`:

````python
abs(-4)
````

Хоча, оскільки функції &mdash; це відповідності, то можна уявити і функцію, що перетворює цілі числа в булеві значення (`True`/`False`). Наприклад таку, що кожному парному числу ставить у відповідність `True`, а непарному &mdash; `False`.

Або ж функцію, що ставить у відповідність кортежу з двох булевих значень інше булеве значення. Наприклад, функція `and` &mdash; вона ставить у відповідність:
- `True`, `True` -> `True`
- `True`, `False` -> `False`
- `False`, `True` -> `False`
- `False`, `False` -> `False`

Хоча в математиці можуть бути і якісь власні складніші функції, як-от
$$
g(x, y) = \begin{cases}
    x^3-2y^2-12x+4, & \text{якщо } x > 0, \\
    |y + 2| + 4, & \text{інакше}
\end{cases}
$$

Функції часто застосовують, аби виконати певну абстракцію (приховати деталі реалізації) і більше про неї не думати. Наприклад, тепер протягом наступних обчислень можна просто писати $g(4, 5)$, і кожного разу не доведеться повторювати цю складну формулу.

## 2. Функції в Python

Попри те, що у Python є багато своїх вбудованих функцій (див. [Built-In functions](https://docs.python.org/3.11/library/functions.html)), зрозуміло, що вони навіть близько не покривають усі необхідні інструменти для написання програм. Тому у Python можна робити свої власні функції.

Створімо для початку функцію, що підносить число до квадрату $f(x)=x^2$. Для цього скористаємося таким синтаксисом:

````python
def f(x):
    result = x**2
    return result
````

Ключове слово `def` (define) означає "визначити". Після нього ми задаємо назву нашої майбутньої функції &mdash; у цьому випадку назва &mdash; `f`.

Одразу за назвою (без пропусків) ставимо дужки, і в них перелічуємо через кому аргументи (параметри) функції. Оскільки функція прийматиме тільки одне число, то пишемо тільки один аргумент &mdash; `x` (хоча могли назвати цей аргумент будь-як).

Після дужок йде двокрапка, а з нового рядка починається "тіло" функції &mdash;  що вона, власне, буде робити. У нашому випадку потрібно знайти квадрат числа `result = x**2`.

Після цього ми визначаємо, що функція буде повертати (що вона поставить у відповідність значенню на вході) з допомогою `return <значення>`.

Однак саме визначення функції ще нічого не робить. Так само, як після визначення $f(x)=x^2$ потрібно виконати $f(2)$, аби дістати `4`, так само й тут потрібно виконати `f(2)`, аби отримати `4`:

````python
print(f(2))
print(f(-3))
````

Можна також провести аналогію між викликом функції та викликом кожного її рядка. Наприклад, якщо виконати такий рядок:

````python
f(4)
````

То це те ж, що підставити `x = 4` і виконати тіло функції:

````python
x = 4
result = x**2
result
````

Як і з математичними функціями, можна задавати різну кількість параметрів на вхід: одну, дві чи навіть нуль:

````python
def func_with_two_args(x, y):
    average = (x + y) / 2
    return f"{x = }, {y = }, {average = }"

def func_with_zero_args():
    return "Немає аргументів"

print(func_with_two_args(4, 5))
print(func_with_zero_args())
````

`Примітка`: Є різниця між поняттями параметрів та аргументів функції. Параметри визначаються іменами, які з'являються у визначенні функції, тоді як аргументи - це значення, які фактично передаються у функцію під час її виклику. Параметри визначають, які аргументи може приймати функція. Наприклад, для функції `func_with_two_args` обʼєкти `x` та `y` є параметрами. Але при виклику функції `func_with_two_args(4, 5)` значення 4 та 5 є аргументами. Часто ці два поняття використовують синонімічно чим ми теж періодично будемо грішити.

## 3. Процедури

Функції можуть нічого не повертати (а точніше, в Python, повертати `None`). Такі функції часто називають **процедурами**. Їх, наприклад, використовують аби щось донести до користувача, як-от через `print`:

````python
def func_with_two_args(x, y):
    average = (x + y) / 2
    print(f"{x = }, {y = }, {average = }")

def func_with_zero_args():
    print("Немає аргументів")

func_with_two_args(4, 5)
func_with_zero_args()
````

`return` немає, тож Python сам повертає `None` у будь-якому випадку.

Але це також означає, що якщо спробувати вивести результат виконання процедур, то вкінці виведеться `None`:

````python
print(func_with_zero_args())
````

Якщо ваша функція виконує дії, але не має чіткого і корисного значення, що повертається (власне є процедурою), ви можете не повертати значення None, оскільки це буде просто зайвим та погіршить сприйняття коду.

## 4. Назви функцій

Назви функцій, так само як і назви змінних, мають бути змістовними. У математиці ми зазвичай бачимо назви функцій достатньо короткі (одна літера!). Це обумовлено тим, що мета у математики  &mdash; створити модель ситуації, тобто абстрагуватися та вирішувати узагальнену задачу. Ба більше, зазвичай в математиці працюють одночасно з невеликою кількістю різнорідних обʼєктів, на відміну від завдань для програмування, тому довгі назви лише б забирали час та робили б міркування громіздкими до сприйняття.

У програмуванні ж навпаки &mdash; назви змінних дуже важливі, адже важливо розуміти з якими обʼєктами ми працюємо в моменті. Ба більше, розробники зазвичай пишуть у середовищах із підтримкою автодоповнення: якщо створити змінну `very_long_name` і почати писати `very_`, то програма одразу підкаже доповнення, що Ви скоріше за все хотіли написати цілу змінну `very_long_name`, і ви зможете прийняти це доповнення з допомогою `Tab` чи `Enter`.

Тому настійливо рекомендуємо давати функціям, так само як і змінним, імена, що пояснюють їхню роль. Тож функцію підняття до квадрату ми могли би назвати `square`, а аргумент &mdash;  `number`:

````python
def square(number):
    result = number**2
    return result

square(2)
````

Варто називати функції за допомогою дієслів аби пояснювати, що вони роблять (бо вони ж виконують функцію!). Дуже рідко можна назвати функцію за допомогою іменника з акцентом на те, що повертають (наприклад, коли будується якийсь складний обʼєкт). Наприклад, для стрічок ми бачимо такі методи як: `capitalize`, `isalpha`, `find`, що одразу свідчать про те, що буде виконувати цей метод.

## 5. Декілька `return`

Можемо тепер також спробувати реалізувати нашу складнішу функцію:
$$
g(x, y) = \begin{cases}
    x^3-2y^2-12x+4, & \text{якщо } x > 0, \\
    |y + 2| + 4, & \text{інакше}
\end{cases}
$$

````python
def complicated_function(x, y):
    if x > 0:
        return x**3 - 2 * y**2 - 12*x + 4
    else:
        return abs(y + 2) + 4

print(f"{complicated_function(1, 2) = }")
print(f"{complicated_function(-1, 2) = }")
````

Як бачимо, можна мати декілька рядків з виразами, що повертає функція. Однак вони всі взаємно виключні: якщо виконався `if`, то повернувся перший вираз &mdash; інакше другий. Як тільки виконується інструкція з ключовим слово `return` виконання тіла функції припиняється (окрім цікавого випадку з try-except, що покажемо далі). Тобто функція не може повернути декілька разів за раз, що абсолютно співпадає з математикою: має результатом бути лише одне значення.

Наприклад:

````python
def multiple_returns(number):
    return number**2
    return number**3

multiple_returns(2)
````

Це також означає, що якщо маємо певний `if`-блок, у якому відбувається `return`, то після цього блоку не потрібно ставити `else` &mdash; код знизу навіть не виконається, якщо станеться повернення з функції! Тож дві наступні функції повністю однакові:

````python
def absolute_value_1(number):
    if number > 0:
        return number
    else:
        return -number

def absolute_value_2(number):
    if number > 0:
        return number
    return -number


print(absolute_value_1(4), absolute_value_2(4))
print(absolute_value_1(-5), absolute_value_2(-5))
````

Рекомендовано використовувати другий варіант, оскільки тоді відсутній додатковий відступ для блоку `else`, і читабельність покращується.

Якщо так переписати нашу функцію, то вийде:

````python
def complicated_function(x, y):
    if x > 0:
        return x**3 - 2 * y**2 - 12*x + 4
    return abs(y + 2) + 4

print(f"{complicated_function(1, 2) = }")
print(f"{complicated_function(-1, 2) = }")
````

Щодо `try-except-finally`: коли `return` передає керування до інструкції `try` з оператором `finally`, то саме його інструкція виконується до того, як функція дійсно вийде з виконання.

````python
def try_finally_return(number):
    try:
        return number**2
    except:
        print('except')
    finally:
        print('finally')

print(try_finally_return(2))
````

## 6. Абстракція та вбудовані функції

Функції дуже зручні для абстракції певних частинок коду.

Наприклад, якщо в нашому коді потрібно буде повторювати дію, яку виконує `complicated_function`, кілька разів, то потрібно буде кілька разів викликати функцію і не думати про те, як вона всередині працює:

````python
print(f"{complicated_function(1, 2) = }")
print(f"{complicated_function(-1, 2) = }")
print(f"{complicated_function(-1, -2) = }")
````

До того ж вони дозволяють не повторювати код кілька разів. У цьому випадку не писати весь код перевірки й обчислень кожен раз, як тут:

````python
if 1 > 0:
    print(f"{1**3 - 2 * 2**2 - 12*1 + 4 = }")
else:
    print(f"{abs(2 + 2) + 4} = ")

if -1 > 0:
    print(f"{-1**3 - 2 * 2**2 - 12*-1 + 4 = }")
else:
    print(f"{abs(2 + 2) + 4 = }")

if -1 > 0:
    print(f"{-1**3 - 2 * -2**2 - 12*-1 + 4 = }")
else:
    print(f"{abs(-2 + 2) + 4 = }")
````

Так само багато речей, які робляться складно, абстрагували від нас і дали вбудовані функції, з багатьма з яких ми вже знайомі.

Той же `print` теж всередині реалізований складніше, але ми маємо можливість його використовувати із просто передаванням стрічок:

````python
print("42")
````

І не тільки стрічок:

````python
print(42)
print({42: 0.6931})
````

Або `max`. Її, з допомогою циклів і розгалужень, і ви вже змогли би написати, але вона зручно абстрагована і є вбудованою:

````python
max({1, 5, 3, 12, 42, 7})
````

## 7. Простори імен

Ми бачили, що якщо в дужках задати назви аргументів через кому, то їх можна буде, як і будь-яку іншу змінну, використовувати всередині функції. Також усередині функції можна створювати нові змінні і використовувати їх. Однак ці всі змінні тимчасові й після виконання функції вже непотрібні.

Наприклад, для підрахунку коренів квадратного рівняння може знадобитися зробити додаткові змінні:

````python
def get_roots(a, b, c):
    sqrt_of_discriminant = (b**2 - 4*a*c)**0.5
    first_root = (-b + sqrt_of_discriminant) / (2 * a)
    second_root = (-b - sqrt_of_discriminant) / (2 * a)
    return (first_root, second_root)

print(get_roots(1, 4, 3))
````

Після виконання функції змінні `a`, `b`, `c`, `sqrt_of_discriminant`, `first_root` та `second_root` вже не потрібні!

Тому в Python, як і в багатьох інших мовах, є поняття **простору імен** &mdash; такого контейнера з іменами усіх об'єктів. Поки ми працювали тільки з **глобальним** простором імен &mdash; його змінні доступні всюди протягом виконання програми.

Однак при запуску функції створюється **локальний** простір імен &mdash; лише для цього виклику. Зсередини функції можна доступитися до *глобального* простору, однак ззовні функції до її *локального* &mdash; ні.

Дуже гарно це можна побачити за допомогою [візуалізатора](https://pythontutor.com/render.html#mode=display), якщо закинете приклад зверху.


Тож якщо ми зараз спробуємо доступитися до, наприклад, `first_root`, то отримаємо помилку, адже ця змінна існувала лише всередині функції під час її виконання:

````python
print(first_root)
````

Однак, якщо ми спробуємо доступитися до *глобальних* змінних зсередини функції, то все вийде:

````python
lyrics = "We're no strangers to love"

def access_global():
    print("У функції:", lyrics)

access_global()
print("Поза функцією:", lyrics)
````

Проте *переприсвоювати* глобальні змінні зсередини функцій не вийде:

````python
lyrics = "We're no strangers to love"

def access_global():
    lyrics = "You know the rules"
    print("У функції:", lyrics)

access_global()
print("Поза функцією:", lyrics)
````

Як тільки ми намагаємося переприсвоїти значення, то змінна стає *локальною* &mdash; тоді водночас існує дві змінні:
- глобальна `lyrics`;
- локальна `lyrics` &mdash; діє лише всередині функції та "перекриває" глобальну на певний час.

Так само, якщо ми захочемо змінити глобальну змінну `lyrics` через `+=`, то не зможемо:

````python
lyrics = "We're no strangers to love"

def access_global():
    lyrics += "\nYou know the rules"
    print("У функції:", lyrics)

access_global()
print("Поза функцією:", lyrics)
````

Оскільки ми намагаємось змінити `lyrics`, то інтерпретатор вважає, що  змінна `lyrics`є *локальною* змінною, що перекриває *глобальну*. Після цього до (ще не існуючої *локальної*) стрічки ми намагаємося додати текст &mdash; виникає помилка.

## 8. `global`

Якщо дуже захотіти, то можна використати синтаксис `global lyrics`, аби вказати Python, що ми дійсно хочемо змінити *глобальну* змінну зсередини функції:

````python
def print_changed_lyrics():
    global lyrics
    lyrics = "Haha, no lyrics"
    print("У функції:", lyrics)
````

````python
lyrics = "We're no strangers to love"

print_changed_lyrics()
print("Поза функцією:", lyrics)
````

Однак таке використання глобальних змінних зазвичай вважаються поганою практикою програмування, адже така поведінка може збивати з пантелику і раптово змінювати значення глобальних змінних. Якщо просто поглянути на код у блоці вище, то нічого не підказує, що `lyrics` раптом зміниться після цієї процедури.

Тож краще ***не використовувати*** ключового слова `global` і розуміти всі його наслідки.

## 9. Іменовані параметри

Ми бачили, що при виклику функції, перше передане значення піде в перший аргумент, друге &mdash; в другий, і так з усіма. Однак деколи може бути зручніше явно задавати, які параметри ми передаємо.

Наприклад, у випадку `get_roots` можна явно задати назви параметрів (через `<назва>=<значення>`), і тоді їхній порядок не важливий:

Є ще й інше застосування іменованих аргументів: параметри із значеннями за замовчуванням. Для цього необхідно при визначенні функції задати значення за замовчуванням через дорівнює:

````python
def join_path(first_part, second_part, os_type="linux"):
    print(f"{os_type = }")
    if os_type in {"linux", "macos", "bsd"}:
        return f"{first_part}/{second_part}"
    return f"{first_part}\\{second_part}"
````

## 11. Змінна кількість параметрів

Так само, як в одну змінну можна присвоювати решту зі змінної кількості значень:

````python
x, *y = 1, 2, 3, "4", {5}

print(f"{x = }")
print(f"{y = }")
````

Так можна і приймати змінну кількість аргументів:

````python
def variable_args(x, *y):
    print(f"{x = }")
    print(f"{y = }")

variable_args(1, 2, 3, "4", {5})
````

Однак, на відміну від присвоєння змінної кількості значень, вони потріпляють у кортеж, а не список, а також не можна використовувати таку конструкцію:

````python
x, *y, z = 1, 2, 3, "4", {5}
print(f"{x = }")
print(f"{y = }")
print(f"{z = }")

def variable_args(x, *y, z):
    print(f"{x = }")
    print(f"{y = }")
    print(f"{z = }")

variable_args(1, 2, 3, "4", {5})
````

У функціях аргумент з `*` може бути тільки один, і він має бути останнім. Після нього всі аргументи вважаються іменованими (ключовими). Тож тепер `z` можна передати лише так:

````python
variable_args(1, 2, 3, "4", z={5})
````

Такі змінні кількості аргументів, наприклад, можна використати для нашої попередньої задачі про з'єднання шляхів файлової системи, аби був не тільки початок і кінець, а й кілька шляхів посередині:

````python
def join_paths(first_part: str, *rest_parts: str, os_type: str = "linux") -> str:
    print(f"{first_part = }")
    print(f"{rest_parts = }")
    print(f"{os_type = }")
    if os_type in {"linux", "macos", "bsd"}:
        delimiter = "/"
    else:
        delimiter = "\\"
    path = first_part
    for part in rest_parts:
        path += delimiter + part
    return path

join_paths("/tmp", "pip-metadata-7j7zc715", "METADATA", os_type="linux")
````

Однак є також можливість отримати всі ключові аргументи, які б користувач не передав, як один. Тоді використовують `**`:

````python
def variable_kwargs(x, **y):
    print(f"{x = }")
    print(f"{y = }")

variable_kwargs(1, a=2, b=3, c="4", d={5})
variable_kwargs(a=2, b=3, c="4", d={5}, x=1)
````

Тоді всі звичайні позиційні та іменовані аргументи передадуться у відповідні змінні, а решта іменованих аргументів буде зібрана у словник.

Часто цим змінним, що збирають решту аргументів, дають спеціальні імена: `*args` (arguments - аргументи) та `**kwargs` (keyword arguments - іменовані/ключові аргументи):

````python
def many_variables(*args, **kwargs):
    print(args)
    print(kwargs)

many_variables(1, 2, a=3, b=4)
````

## 13. Документація

Розібравшись, як працюють функції, також потрібно зрозуміти, як їх правильно писати. Ви напевно вже стикалися, що часто не виходить просто взяти і одразу написати код правильно і зрозуміло. Тож варто постійно мати орієнтир, куди ми прямуємо.

Тож спочатку варто визначитися із тим, що функція має виконувати (зазвичай, вам уже дається умова задачі), і як із нею взаємодіяти (що функція приймає та що повертає). Після цього має бути готовий хоча б рядок визначення функції, з аргументами, анотаціями типів аргументів та значенням, яке функція повертає, як у цьому прикладі:

```py
def get_hypotenuse(a: float, b: float) -> float:
```

У цьому випадку ми вирішилися, що функція має приймати два числа типу `float` і повертати число типу `float`. Назва функції ж пояснює, що вона має робити.

Але згадаймо злегка складніший приклад з'єднання шляхів файлової системи:

```py
def join_path(first_part: str, second_part: str, os_type: str = "linux") -> str:
```

Якщо про роль `first_part` та `second_part` ще можна здогадатися, то які значення приймає `os_type` із самого лишень визначення складно здогадатися. Якщо ж ви пам'ятаєте з частини про анотації, то при використанні функцій тільки визначення і показує:



Тож для майбутніх користувачів (в тому числі нас самих через кілька місяців, коли вже не пам'ятатимемо, як саме все працює) варто якось пояснити, як користуватися нашим творінням.

Для цього використовують документацію. Звичайно, можна в окремому файлі писати пояснення до кожної функції, але постійно переключатися між файлами не дуже зручно. Натомість у Python є підтримувана можливість писати документацію прямо в коді. Це робиться за допомогою рядків, що починаються з трьох подвійних лапок `"""` або `'''` (одразу після першого рядка функції) і закінчуються також трьома подвійними лапками:

````python
def join_path(first_part: str, second_part: str, os_type: str="linux"):
    """Join two parts of the path using the specified delimiter.

    :param first_part: The start of the path.
    :param second_part: The end of the path.
    :param os_type: The operating system, on which the path will be used. Should be
        from the list {"linux", "macos", "bsd", "windows"}. Default is "linux".
    :return: The joined path with the delimiter appropriate for the specified operating system.
    """
    if os_type in {"linux", "macos", "bsd"}:
        return f"{first_part}/{second_part}"
    return f"{first_part}\\{second_part}"
````

Їх часто називають **докстрічками** (docstrings) чи стрічками документації, і в них, зазвичай, спершу пишуть пояснення, що власне робить функція. Після цього для кожного аргумента (параметра) йде окреме пояснення, яка його роль. Так, у відділі для `os_type` можемо пояснити, які значення цей аргумент приймає та на що це впливає. Врешті пишуть, що функція повертає.

Тепер, якщо ми спробуємо скористатися функцією:



Якщо використовувати стиль позначень параметрів та результату, як показано вище, то автодоповнення навіть зрозуміє суть та гарно все оформить, зі списком та підсвіткою.

**Примітка**:
> Докстрічки, як і назви змінних чи функцій, прийнято писати англійською мовою. Якщо ж у вас із нею труднощі, то рекомендуємо скористатися перекладачем чи мовною моделлю для перекладу.

## 14. `doctest`

Однак часто недостатньо просто визначитися, що має виконувати функція та які її входи та виводи. Деколи задачі доволі складні (згадаймо "П'яну вишню"!), і в них легко помилитися &mdash; тож хотілося би бути впевненими, що при всіх можливих випадках функція працює правильно.

Для такого буває зручно написати інший код, який тестує, чи правильно працює цей код &mdash; адже часто перевірити правильність легше, ніж повністю написати всю логіку.

Такі тести, які перевіряють окремі частинки коду (зазвичай функції), називають **модульними тестами** (Unit tests).

Одним із найпростіших способів так тестувати окремі функції &mdash; це вбудований модуль `doctest`. Якщо його використовувати, то додається спеціальний синтаксис в докстрічках - дуже схожий на роботу в інтерактивному Python:

````python
PI = 3.14

def circle_area(radius: float) -> float:
    """Find the area under a circle.

    The approximation for PI is 3.14.

    >>> circle_area(2)
    12.56
    >>> circle_area(1)
    3.14
    """
    return PI * radius**2
````

Після `>>> ` можна писати код, який ми тестуємо, а в рядку під ним те, що має бути результатом виконання. А сама функція досить проста - обрахунок площі круга за формулою $S=\pi \cdot r^2$

На перший погляд нічого не змінилося. Однак тепер, якщо використати `doctest`:

````python
import doctest
print(doctest.testmod())
````

То виявиться, що всі тести пройшли. Однак тільки що ми пропустили виняткові випадки (corner case, edge case): якщо радіус від'ємний!

Навіть на такому простому прикладі, як виявилося, легко допустити помилку, якщо не подумати, які можуть бути такі незвичайні випадки. У складніших же ситуаціях пропустити якусь умову, не подумати про якусь неочікувану поведінку користувача чи просто щось забути врахувати набагато легше, а потім знайти такі помилки набагато важче.

Саме тому від початку краще подумати над винятковими випадками, що з ними робити і написати кілька тестів, аби потім не забути.

Тоді, наприклад, ми би подумали: а що, як цей `float`, який нам передають, якийсь не такий, як ми очікуємо (оскільки ми знаємо, що це `float`, то не потрібно думати про якісь інші випадки, де передають взагалі не число).

Тоді ми би могли подумати за якісь дуже великі числа &mdash; але з ними, здається, програма мала би так само працювати, як і з малими.

Також про нуль &mdash; на нього не можна ділити, при множенні він все інше в нуль перетворює, тож, може, він може негативно впливати. Але ми не використовуємо ніде ділення (те, чи коло з радіусом `0` є колом &mdash; філософське питання. Однак нехай ми вирішилися, що це таки коло, просто точкове).

Однак що робити з від'ємними радіусами? Радіус &mdash; це за замовчуванням відстань, а мати від'ємні відстані якось зовсім нелогічно. Тож, як один з варіантів, можемо просто нічого не повертати (`None`), якщо передають від'ємний радіус. Додамо це до тестів (оскільки в інтерактивному Python при `>>> None` нічого не виводиться, то й тут потрібно лишити порожній рядок)!

````python
PI = 3.14

def circle_area(radius: float) -> float:
    """Find the area under a circle.

    The approximation for PI is 3.14.

    >>> circle_area(2.0)
    12.56
    >>> circle_area(1.0)
    3.14
    >>> circle_area(0.0)
    0.0
    >>> circle_area(-1)

    """
    return PI * radius**2

import doctest
print(doctest.testmod())
````

Оскільки ми не додали функціоналу для відсіювання від'ємних радіусів, то тест не проходить. При цьому `doctest` явно підказує, який тест не пройшов, який був очікуваний результат ("Expected Nothing") та який насправді результат ("Got ...").

Тож довершімо цю функцію, аби вона повертала `None`:

````python
PI = 3.14

def circle_area(radius: float) -> float:
    """Find the area under a circle.

    The approximation for PI is 3.14.

    >>> circle_area(2.0)
    12.56
    >>> circle_area(1.0)
    3.14
    >>> circle_area(0.0)
    0.0
    >>> circle_area(-1)

    """
    if radius < 0:
        return None
    return PI * radius**2

import doctest
print(doctest.testmod())
````

Немає жодного тесту `failed`!

Таким чином, якщо впевнені, що продумали всі незвичайні випадки, то матимемо орієнтир, куди рухатися. Тепер, якщо всі тести проходитимуть, то матимемо хорошу впевненість, що функція працює правильно.

Загалом варто додавати кілька тестів для різних типів очікуваних значень (аби раптом не зламати навіть звичайної поведінки), а потім пошукати виняткові випадки й перевірити їх.

Після цього можна один за одним викреслювати зі списку тести, які код починає проходити, і радіти життю!

`Примітка`: Зрозуміло, що доктести не годяться для серйозного тестування, як з урахуванням читабельності коду так і з точки зору функціоналу. Тому доктести використовують дійсно як частину документації, щоб показати очікувану роботу даної функції. При цьому поки ми вчимося створювати невеличкі функції, то це ідеальний інструмент мати усе в одному місці.
