````python
def get_roots(a, b, c):
    sqrt_of_discriminant = (b**2 - 4*a*c)**0.5
    first_root = (-b + sqrt_of_discriminant) / (2 * a)
    second_root = (-b - sqrt_of_discriminant) / (2 * a)
    return (first_root, second_root)

print(get_roots(1, 4, 3))
````

````python
def join_paths(first_part: str, *rest_parts: str, os_type: str = "linux") -> str:
    print(f"{first_part = }")
    print(f"{rest_parts = }")
    print(f"{os_type = }")
    if os_type in {"linux", "macos", "bsd"}:
        delimiter = "/"
    else:
        delimiter = "\\"
    path = first_part
    for part in rest_parts:
        path += delimiter + part
    return path

join_paths("/tmp", "pip-metadata-7j7zc715", "METADATA", os_type="linux")
````

````python
def join_path(first_part: str, second_part: str, os_type: str="linux"):
    """Join two parts of the path using the specified delimiter.

    :param first_part: The start of the path.
    :param second_part: The end of the path.
    :param os_type: The operating system, on which the path will be used. Should be
        from the list {"linux", "macos", "bsd", "windows"}. Default is "linux".
    :return: The joined path with the delimiter appropriate for the specified operating system.
    """
    if os_type in {"linux", "macos", "bsd"}:
        return f"{first_part}/{second_part}"
    return f"{first_part}\\{second_part}"
````
